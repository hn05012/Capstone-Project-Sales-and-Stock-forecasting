```python
# data_2017 = data.loc['2017-04-01':'2017-12-31']
# data_2018 = data.loc['2018-01-01':'2018-12-31']
# data_2019 = data.loc['2019-01-01':'2019-12-31']
# data_2020 = data.loc['2020-01-01':'2020-12-31']
# data_2021 = data.loc['2021-01-01':'2021-11-30']

categories = []
brands = []
for i in range(1, len(itemlist)):
    if itemlist[i][0] not in categories:
        categories.append(itemlist[i][0])
    if itemlist[i][3] not in brands:
        brands.append(itemlist[i][3])


```

```python
cat_brand_sold = dict()
for c in categories:
    cat_brand_sold[c] = []
for b in brands:
    cat_brand_sold[b] = []

cat_brand_sold['dates'] = []
```

```python
cat_brand_sold_encoded = {}
codes = []
code = ''
for key in cat_brand_sold:
    if key in categories:
        while True:
            code = 'category ' + random.choice(string.ascii_letters) + str(random.randint(0,255)) + random.choice(string.ascii_letters)
            if code not in codes:
                codes.append(code)
                break
            else:
                continue
        cat_brand_sold_encoded[key] = code
    elif key in brands:
        while True:
            code = 'brand ' + random.choice(string.ascii_letters) + str(random.randint(0,255)) + random.choice(string.ascii_letters)
            if code not in codes:
                codes.append(code)
                break
            else:
                continue
        cat_brand_sold_encoded[key] = code

cat_brand_sold_encoded
```

```python
i = 0
for index, row in sales_df.iterrows():

    date = row['Customer Date']
    item_code = row['Design Number']
    quantity_sold = row['Quantity']
    
    if any(item_list_df.ItemCode == item_code):
        category = item_list_df[item_list_df['ItemCode'] == item_code]['Category'].values[0]
        brand = item_list_df[item_list_df['ItemCode'] == item_code]['ItemName'].values[0]

        # checks
        if category not in cat_brand_sold:
            print(category + ' category not there')
        
        if brand not in cat_brand_sold:
            print(brand + ' brand not there')
        

        cat_brand_sold['dates'].append(date)
        cat_brand_sold[category].append(quantity_sold)
        cat_brand_sold[brand].append(quantity_sold)

        for key in cat_brand_sold:
            if key not in ['dates', category, brand]:
                cat_brand_sold[key].append(0)
    else:
        i+=1
        print(item_code)
print(i)



    
```

```python
cat_brand_df = pd.DataFrame.from_dict(cat_brand_sold)    
```

```python
cat_brand_df['dates'] = pd.to_datetime(cat_brand_df['dates'], dayfirst=True)
cat_brand_df.set_index('dates', inplace=True)
```
